```python
%matplotlib qt

from matplotlib.backend_bases import MouseButton
import matplotlib.pyplot as plt
import numpy as np

def on_click(event):
    if event.button is MouseButton.LEFT and event.inaxes:
        mx, my = event.xdata, event.ydata
        ax = event.inaxes
        ax.streamplot(*drs, *dFs, linewidth=1, arrowstyle='fancy', start_points=[[mx,my]])
        ax.figure.canvas.draw()
```

## Aufgabe 7.2

```python
a = 1.75
b = 1.2

def F(x, y):
    return x*(-x - y + 1), \
           y*(-a*x - y + b)

def Fl(x,y):
    return x*(-y + 1), \
           y*(-a*x + b)


### grid ###
n = 20
xmin, xmax = -0.5, 1.5
ymin, ymax = -0.5, 1.5

xi = np.linspace(xmin, xmax, n)
yi = np.linspace(ymin, ymax, n)
rs = np.meshgrid(xi, yi) # Ortsvektoren

Fs = Fl(*rs)

F_abs = np.sqrt(Fs[0]**2 + Fs[1]**2)
fs = Fs/F_abs * np.tanh(5*F_abs) # Skalierung


### plots ###
plt.figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k')

plt.quiver(*rs, *fs, scale=30, alpha=0.4)


### Nullklinen ###
plt.plot(np.zeros_like(yi), yi, '--k', color = "r", label='$x$-Nullkline')
plt.plot(1-yi, yi, '--k', color = "r")

plt.plot(xi, np.zeros_like(xi), ':k', color = "b", label='$y$-Nullkline')
plt.plot(xi, b-a*xi, ':k', color = "b")
plt.legend()


### dense grid ###
N = 100
dxi = np.linspace(xmin,xmax,N)
dyi = np.linspace(xmin,xmax,N)
drs = np.meshgrid(dxi,dyi)

dFs = F(*drs)


### handles ###
plt.connect('button_press_event', on_click)
plt.connect('motion_notify_event', on_click)


plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

plt.show()
```

## ODE Van der Pol

```python
mu = 1.5

def F(x, y):
    return y - mu* (x**3/3 - x), \
           -x


### grid ###
n = 40
xmin, xmax = -5, 5
ymin, ymax = -5, 5

xi = np.linspace(xmin, xmax, n)
yi = np.linspace(ymin, ymax, n)
rs = np.meshgrid(xi, yi) # Ortsvektoren

Fs = F(*rs)

F_abs = np.sqrt(Fs[0]**2 + Fs[1]**2)
fs = Fs/F_abs * np.tanh(20*F_abs) # Skalierung


### plots ###
plt.figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k')

plt.quiver(*rs, *fs, scale=30, alpha=0.4)


def f(x):
    return mu * (x**3 /3 - x)
        

### Nullklinen ###
#plt.plot(np.zeros_like(yi), yi, '--k', color = "r", label='$x$-Nullkline')
#plt.plot(1-yi, yi, '--k', color = "r")

plt.plot(xi, f(xi), ':k', color = "b", label='$y$-Nullkline')
#plt.plot(xi, b-a*xi, ':k', color = "b")
plt.legend()


### dense grid ###
N = 100
dxi = np.linspace(xmin,xmax,N)
dyi = np.linspace(xmin,xmax,N)
drs = np.meshgrid(dxi,dyi)

dFs = F(*drs)


### handles ###
plt.connect('button_press_event', on_click)
plt.connect('motion_notify_event', on_click)


plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

plt.show()
```
